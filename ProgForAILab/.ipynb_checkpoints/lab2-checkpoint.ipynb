## Task 2

2. Announcement Preview
You're developing an announcement preview tool for a messaging platform. The 
tool needs to provide a concise preview of longer messages.

Task:

Given a message: "Join us for the webinar on Aug 30, 2023, 2:00 PM (GMT)", 
perform the following tasks:
* Calculate the length of the message.
* Replace the word "webinar" with "online session".
* Extract the date "Aug 30, 2023".
* Create a shortened preview containing only the first 40 characters.
* Extract time.
Print all the results.

```python
message = "Join us for the webinar on Aug 30, 2023, 2:00 PM (GMT)"
print(message)

print("length of message " , len(message))

message.replace("webinar","online session")
print("The new version ",message)

exact_date = message[27:39]
print(exact_date)

print(message[:40])

print("Time : ", message[40:48])
```
