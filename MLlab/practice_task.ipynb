5. Write a program to input marks of five subjects Physics, Chemistry, Biology,
Mathematics, and Computer. Calculate percentage and grade according to
following: Percentage >= 90% : Grade A
Percentage >= 80% : Grade B
Percentage >= 70% : Grade C
Percentage >= 60% : Grade D
Percentage >= 40% : Grade E
Percentage < 40% : Grade F

```python
subject = ['Physics',"Chemistry","Biology","Mathematics","Computer"]
total_marks = 500
subjects_marks = []
for i in range(len(subject)):
    marks = float(input(f"Enter {subject[i]} marks : "))
    if marks > 100:
        raise ValueError
    subjects_marks.append(marks)
percentage = (sum(subjects_marks) / total_marks ) * 100

print(f"Percentage : {percentage}")
if percentage >= 90:
    print(f"Grade A ")
elif percentage >= 80:
    print("Grade B")
elif percentage >= 70:
    print("Grade C")
elif percentage >= 60:
    print("Grade D")
elif percentage >= 40:
    print("Grade E")
elif percentage <40:
    print("Grade F")
```

```python
# right pascal triangle

rows = 5

for i in range(0,rows):
    for j in range(0 , i+1):
        print("*",end="")
    print("\r")

for i in range(rows , 0, -1):
    for j in range (0 , i -1):
        print("*",end="")
    print("\r")
```

```python
rows = 5

for i in range(0, rows):          # Outer loop: decides the row number (0, 1, 2, 3, 4)
    for j in range(0, i + 1):     # Inner loop: prints '*' based on the current row index
        print("*", end="")        # end="" keeps stars on the same line
    print("\r")                   # Moves the cursor to the start of the next line

for i in range(rows, 0, -1):      # Outer loop: counts backwards (5, 4, 3, 2, 1)
    for j in range(0, i - 1):     # Inner loop: prints stars (i - 1) times
        print("*", end="")
    print("\r")
```

```python
sam_data = "Python 3.2"

Letters = 0
Digits = 0


for char in sam_data:
    if char.isalpha():
        Letters += 1
    elif char.isdigit():
        Digits += 1

print(f"Letters : {Letters}")
print(f"Digits : {Digits}")
```

12. Write a Python program that removes and prints every third number from a list of 
numbers until the list is empty.   

```python
numbers = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# We start at index 2 because it's the 3rd element (0, 1, 2)
position = 2

print(f"Original list: {numbers}")
print("Removing elements:")

while len(numbers) > 0:
    # If the list has fewer elements than our target position,
    # we wrap around using the modulo operator (%)
    position = position % len(numbers)
    
    # Remove and return the item at the calculated position
    removed_element = numbers.pop(position)
    print(removed_element)
    
    # Move the pointer forward by 2 
    # (Since we popped one, the next '3rd' element is 2 indices away)
    position += 2

print("List is now empty.")
```
